```python
#Ask for student's name
name = input("Enter student's name: ")

#Ask for student's age
age = input("Enter student's age: ")

#Ask for student's course
course = input("Enter student's course: ")

#Display formatted output
print("--- Student Information ---")
print("Name:", name)
print("Age:", age)
print("Course:", course)
```

```python
#Accept student's exam score
score = float(input("Enter student's exam score: "))

#Check if passed or failed
if score >= 75:
  print("Passed")
else:
  print("Failed")
```

```python
#Input student's name and score
name = input("Enter student's name: ")
score = float(input("Enter student's score: "))

#Determine letter grade
if 90 <= score <= 100:
  grade = "A"
elif 80 <= score <= 89:
  grade = "B"
elif 70 <= score <= 79:
  grade = "C"
else:
  grade = "F"

#Display result
print("--- Grade Result ---")
print("Name:", name)
print("Score:", score)
print("Letter Grade:", grade)
```
